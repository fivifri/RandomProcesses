# Modeling random processes with a given distribution law

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
```

```python
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 9),
          'axes.labelsize': 'x-large',
          'axes.titlesize':'x-large',
          'xtick.labelsize':'x-large',
          'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
```

- ***func*** is a random process $X = f(t, U)$
- ***$t$*** is time
- ***$U$*** is a discrete random variable
- ***sections*** is sections of a rancom process
- ***start, stop, step*** set the time range

```python
class RandomProcess:
    def __init__(self, start, stop, step, U, sections, func):
        self.arange = np.arange(start, stop, step)
        self.U = U
        self.sections = sections
        self.func = func

    def Modeling(self):    
        _, ax = plt.subplots()
        
        for u in self.U:
            y = self.func(u, self.arange)
            ax.plot(self.arange, y, label=f"U={round(u, 3)}")

        for u in self.U:
            y = self.func(u, self.sections)
            ax.scatter(self.sections, y)

        for t in self.sections:
            ax.axvline(x=t, linestyle='--', color='gray')
        
        ax.set(xlabel='time', ylabel='process')
        ax.legend()
        
        plt.show()
```

```python
myProcess = RandomProcess(0, 8, 0.1, np.array([0, 0.5, 3**0.5/2]), np.array([0, 1, 3.5, 7]), lambda u, t: t/np.arccos(u))
```

```python
myProcess.Modeling()
```
