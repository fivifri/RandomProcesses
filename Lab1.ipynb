# Modeling random processes with a given distribution law

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class RandomProcess:
    def __init__(self, start, stop, step, U, sections, func):
        self.arange = np.arange(start, stop, step);
        self.U = U
        self.sections = sections
        self.func = func

    def modeling(self):    
        fig, ax = plt.subplots(figsize=(15, 9))
        
        for u in self.U:
            y = self.func(u, self.arange)
            ax.plot(self.arange, y, label=f"U={u}")

        for u in self.U:
            y = self.func(u, self.sections)
            ax.scatter(self.sections, y)

        for t in self.sections:
            ax.axvline(x=t, linestyle='--')

        ax.legend()
        plt.show()
```

```python
myProcess = RandomProcess(0, 8, 0.1, np.array([0, 0.5, 3**0.5/2]), np.array([0, 1, 3.5, 7]), lambda u, t: t/np.arccos(u))
```

```python
myProcess.modeling()
```
